```python
from docx import Document
 # Crear un nuevo documento Word
doc = Document()
 # Crear una tabla con 5 filas y 5 columnas
table = doc.add_table(rows=6, cols=6)
 # Agregar encabezamientos a la primera fila
header_cells = table.rows[0].cells
header_cells[0].text = "Encabezado 1"
header_cells[1].text = "Encabezado 2"
header_cells[2].text = "Encabezado 3"
header_cells[3].text = "Encabezado 4"
header_cells[4].text = "Encabezado 5"
 # Agregar datos a las filas restantes
for i in range(1, 6):
    row_cells = table.rows[i].cells
    row_cells[0].text = "Dato {}-1".format(i)
    row_cells[1].text = "Dato {}-2".format(i)
    row_cells[2].text = "Dato {}-3".format(i)
    row_cells[3].text = "Dato {}-4".format(i)
    row_cells[4].text = "Dato {}-5".format(i)
 # Guardar el documento Word
doc.save("tabla_ejemplo1.docx")
```

```python
from docx import Document
 # Listas de ejemplo
lista1 = ["Elemento 1", "Elemento 2", "Elemento 3"]
lista2 = ["Elemento A", "Elemento B", "Elemento C"]
lista3 = ["Elemento X", "Elemento Y", "Elemento Z"]
lista4 = ["Elemento 10", "Elemento 20", "Elemento 30"]
lista5 = ["Elemento I", "Elemento II", "Elemento III"]
 # Crear un nuevo documento Word
doc = Document()
 # Crear una tabla con 4 filas y 2 columnas (incluyendo el encabezamiento)
table = doc.add_table(rows=6, cols=2)
 # Agregar encabezamientos a la primera fila
header_cells = table.rows[0].cells
header_cells[0].text = "Lista 1"
header_cells[1].text = "Lista 2"
 # Agregar datos a las filas restantes
for i in range(1, 6):
    row_cells = table.rows[i].cells
    row_cells[0].text = ", ".join(lista1)
    row_cells[1].text = ", ".join(lista2)
    lista1, lista2 = lista2, lista3  # Rotar las listas para cada fila
 # Guardar el documento Word
doc.save("tabla_ejemplo.docx")
```

```python
from docx import Document
 # Crear un nuevo documento de Word
doc = Document()
 # Tus listas de datos
lista1 = [1, 2, 3, 4]
lista2 = ['a', 'b', 'c', 'd']
lista3 = [True, False, True, False]
 # Crear una tabla con el número de columnas igual al número de listas
num_columnas = 3
tabla = doc.add_table(rows=1, cols=num_columnas)
 # Agregar encabezamientos a la primera fila de la tabla
encabezamientos = ['Encabezamiento 1', 'Encabezamiento 2', 'Encabezamiento 3']
for i, encabezamiento in enumerate(encabezamientos):
    tabla.cell(0, i).text = encabezamiento
 # Agregar los datos de cada lista en columnas separadas
listas = [lista1, lista2, lista3]
for i, lista in enumerate(listas):
    for dato in lista:
        # Agregar una nueva celda a la columna
        tabla.cell().text = str(dato)
 # Guardar el documento de Word
doc.save("tabla.docx")
```

```python
from docx import Document
 # Crear un nuevo documento de Word
doc = Document()
 # Tus listas de datos
lista1 = [1, 2, 3, 4]
lista2 = ['a', 'b', 'c', 'd']
lista3 = [True, False, True, False]
 # Crear una tabla con el número de columnas igual al número de listas
num_columnas = 3
tabla = doc.add_table(rows=1, cols=num_columnas)
 # Agregar encabezamientos a la primera fila de la tabla
encabezamientos = ['Encabezamiento 1', 'Encabezamiento 2', 'Encabezamiento 3']
for i, encabezamiento in enumerate(encabezamientos):
    tabla.cell(0, i).text = encabezamiento
 # Agregar los datos de cada lista en columnas separadas
listas = [lista1, lista2, lista3]
for i, lista in enumerate(listas):
    for j, dato in enumerate(lista):
        # Agregar una nueva celda a la columna
        celda = tabla.cell(j+1, i)
        celda.text = str(dato)
 # Guardar el documento de Word
doc.save("tabla.docx")
```
