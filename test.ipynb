```python
def encontrar_parejas(lista, valor):
    parejas = []
    diferencias = []

    for num in lista:
        if num < valor:
            parejas.append(num)
        elif num > valor and parejas:
            diferencias.extend([parejas[-1] - x for x in parejas[:-1]])
            parejas = []

    return diferencias

# Ejemplo de uso
lista = [0, 6, 12, 18, 48]
valor = 15

diferencias = encontrar_parejas(lista, valor)

print("Diferencias:")
for diferencia in diferencias:
    print(diferencia)
```

```python
def encontrar_valor_menor_cercano(lista, valor):
    menor_cercano = min(lista, key=lambda x: abs(x - valor))
    indice_menor_cercano = lista.index(menor_cercano)
    operaciones = []

    for i in range(indice_menor_cercano, 0, -1):
        operaciones.append(lista[i] - lista[i - 1])

    return operaciones[::-1]

# Ejemplo de uso
lista_valores = [0, 6, 12, 18, 48]
valor_determinado = 15

resultado = encontrar_valor_menor_cercano(lista_valores, valor_determinado)
resultado.insert(0, valor_determinado - lista_valores[lista_valores.index(resultado[0]) - 1])
print(resultado)
```

```python
def obtener_maximo_menor(lista, valor):
    maximo_menor = max(filter(lambda x: x <= valor, lista))
    return maximo_menor

# Ejemplo de uso
lista_valores = [0, 6, 12, 18, 48]
valor_determinado = 15

resultado = obtener_maximo_menor(lista_valores, valor_determinado)
print(resultado)
```
