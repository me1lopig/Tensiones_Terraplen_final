```python
def insertar_valor(lista, valor):
    lista.append(valor)  # Agrega el valor a la lista
    lista.sort()         # Ordena la lista en orden ascendente
    return lista


# Ejemplo de uso
lista_original = [0, 6, 18, 48]
valor_insertar = 12

lista_ordenada = insertar_valor(lista_original, valor_insertar)
print(lista_ordenada)


insertar_valor(cotas,12)
```

```python
def obtener_maximo_menor(lista, valor):
    maximo_menor = max(filter(lambda x: x <= valor, lista))
    return maximo_menor

# Ejemplo de uso
lista_valores = [0, 5, 12, 18, 48]
valor_determinado = 19

resultado = obtener_maximo_menor(lista_valores, valor_determinado)

print(valor_determinado-resultado)
for j in range(lista_valores.index(resultado),0,-1):
    print (lista_valores[j]-lista_valores[j-1])

# se obtienen los espesores de los estrato desde la cota que se introduce como dato
```
